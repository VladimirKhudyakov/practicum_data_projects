````python
/# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π –≥—Ä–∞–Ω–∏—Ü—ã (–ª–µ–≤–æ–µ –∏ –ø—Ä–∞–≤–æ–µ "–∫—Ä–∞–π–Ω–∏–µ" –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö) [–Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞–∑–º–∞—Ö–∞]
# –ø–∞—Ä–∞–º–µ—Ç—Ä 'feature' - –ø—Ä–∏–∑–Ω–∞–∫ (—Å—Ç–æ–ª–±–µ—Ü –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞)
def left_right_extreme_data_boundary(feature):
    
    if(feature.empty):
        return np.nan, np.nan
    
    # –≤—ã—á–∏—Å–ª–∏–º –∫–≤–∞–Ω—Ç–∏–ª–∏ q3 (75%-–π) –∏ q1 (25% -–π)
    q3, q1 = np.percentile(feature, [75,25])
    
    # –≤—ã—á–∏—Å–ª–∏–º –º–µ–∂–∫–≤–∞–Ω—Ç–∏–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª iqr
    iqr = q3 - q1

    # —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –ª–µ–≤—É—é –∏ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É (–ª–µ–≤–æ–µ –∏ –ø—Ä–∞–≤–æ–µ "–∫—Ä–∞–π–Ω–∏–µ" –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö)
    return np.around((q1 - iqr * 1.5), decimals=3), np.around((q3 + iqr * 1.5), decimals=3)



# —Ñ—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ–¥–Ω–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ
def outlier_handling_in_feature(dataframe,
                                column_name,
                                min_range_limit,
                                max_range_limit):
    
    # —É–¥–∞–ª–∏–º –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞ –≤—ã–±—Ä–æ—Å—ã 'dataframe', –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ 'column_name', –≤—ã–±—Ä–æ—Å—ã
    dataframe = dataframe[~(dataframe[column_name] < min_range_limit) &
                          ~(dataframe[column_name] > max_range_limit)]
    
    return dataframe



def check_in_outlier(features, column_names):
    
    for column_name in column_names:
        left_border, right_border = left_right_extreme_data_boundary(features[column_name].fillna(0))
        
        if ( (left_border > features[column_name].min()) or
             (right_border < features[column_name].max()) ):
            return False
        
    return True


# —Ñ—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ–¥–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
def outlier_handling(features, column_names):
    
    work = True;
    while(work):
        for column_name in column_names:
        
            left_border, right_border = left_right_extreme_data_boundary(features[column_name].fillna(0))
        
        
            features = outlier_handling_in_feature(features,
                                                   column_name,
                                                   left_border,
                                                   right_border)
        
        if (features.empty or check_in_outlier(features, column_names)):
            break
    
    return features
````

## –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

–í –∫–∞—á–µ—Å—Ç–≤–µ –º–æ–¥–µ–ª–∏ –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è - ML –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–≤–µ—Ä—Ç–æ—á–Ω–æ–π –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏, –±—ã–ª–∞ –≤—ã–±—Ä–∞–Ω–∞ –º–æ–¥–µ–ª—å `ResNet152V2` –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `Keras`.




–ò—Å—Ö–æ–¥–Ω—ã–π –∫–æ–¥ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ GPU –≤ —Ç—Ä–µ–Ω–∞–∂–µ—Ä–µ.

```python

# –∏—Å—Ö–æ–¥–Ω—ã–π –∫–æ–¥ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ GPU (–≤ —Ç—Ä–µ–Ω–∞–∂—ë—Ä–µ)
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.resnet_v2 import ResNet152V2


SRC_PATH = '/datasets/faces/'
IMAGES_SUBPATH = 'final_files/'
DATASETS_FILE_NAME = 'labels.csv'
BATCH_SIZE = 16
X_COL = 'file_name'
Y_COL = 'real_age'

IMAGE_HEIGHT = 360
IMAGE_WIDTH = 360
# height, width - images
TARGET_SIZE = (IMAGE_HEIGHT, IMAGE_WIDTH)
CLASS_MODE = 'raw'
RANDOM_SEED = 753


# —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π –≥—Ä–∞–Ω–∏—Ü—ã (–ª–µ–≤–æ–µ –∏ –ø—Ä–∞–≤–æ–µ "–∫—Ä–∞–π–Ω–∏–µ" –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö) [–Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞–∑–º–∞—Ö–∞]
# –ø–∞—Ä–∞–º–µ—Ç—Ä 'feature' - –ø—Ä–∏–∑–Ω–∞–∫ (—Å—Ç–æ–ª–±–µ—Ü –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞)
def left_right_extreme_data_boundary(feature):
    
    if(feature.empty):
        return np.nan, np.nan
    
    # –≤—ã—á–∏—Å–ª–∏–º –∫–≤–∞–Ω—Ç–∏–ª–∏ q3 (75%-–π) –∏ q1 (25% -–π)
    q3, q1 = np.percentile(feature, [75,25])
    
    # –≤—ã—á–∏—Å–ª–∏–º –º–µ–∂–∫–≤–∞–Ω—Ç–∏–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª iqr
    iqr = q3 - q1

    # —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –ª–µ–≤—É—é –∏ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É (–ª–µ–≤–æ–µ –∏ –ø—Ä–∞–≤–æ–µ "–∫—Ä–∞–π–Ω–∏–µ" –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö)
    return np.around((q1 - iqr * 1.5), decimals=3), np.around((q3 + iqr * 1.5), decimals=3)



# —Ñ—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ–¥–Ω–æ–º –ø—Ä–∏–∑–Ω–∞–∫–µ
def outlier_handling_in_feature(dataframe,
                                column_name,
                                min_range_limit,
                                max_range_limit):
    
    # —É–¥–∞–ª–∏–º –∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞ –≤—ã–±—Ä–æ—Å—ã 'dataframe', –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ 'column_name', –≤—ã–±—Ä–æ—Å—ã
    dataframe = dataframe[~(dataframe[column_name] < min_range_limit) &
                          ~(dataframe[column_name] > max_range_limit)]
    
    return dataframe



def check_in_outlier(features, column_names):
    
    for column_name in column_names:
        left_border, right_border = left_right_extreme_data_boundary(features[column_name].fillna(0))
        
        if ( (left_border > features[column_name].min()) or
             (right_border < features[column_name].max()) ):
            return False
        
    return True


# —Ñ—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ–¥–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ
def outlier_handling(features, column_names):
    
    work = True;
    while(work):
        for column_name in column_names:
        
            left_border, right_border = left_right_extreme_data_boundary(features[column_name].fillna(0))
        
        
            features = outlier_handling_in_feature(features,
                                                   column_name,
                                                   left_border,
                                                   right_border)
        
        if (features.empty or check_in_outlier(features, column_names)):
            break
    
    return features

def load_train(path=SRC_PATH):

    train_datagen = ImageDataGenerator(width_shift_range=0.2,
                                       height_shift_range=0.2,
                                       horizontal_flip=True,
                                       shear_range=0.2,
                                       zoom_range=0.2,
                                       rescale=1/255.,
                                       validation_split=0.25)
                                       
    df = pd.read_csv(path + DATASETS_FILE_NAME)
    df = df[df['real_age'] >= 5]
    df = outlier_handling(df, ['real_age'])
         
    train_datagen_flow =\
        train_datagen.flow_from_dataframe(dataframe=df,
                                          directory=path + IMAGES_SUBPATH,
                                          x_col=X_COL,
                                          y_col=Y_COL,
                                          target_size=TARGET_SIZE,
                                          batch_size=BATCH_SIZE,
                                          class_mode=CLASS_MODE,
                                          subset='training',
                                          shuffle=True,
                                          seed=RANDOM_SEED)

    return train_datagen_flow


def load_test(path=SRC_PATH):

    valid_datagen = ImageDataGenerator(validation_split=0.25, rescale=1/255.)
       
    valid_datagen_flow =\
        valid_datagen.flow_from_dataframe(dataframe=pd.read_csv(path + DATASETS_FILE_NAME),
                                          directory=path + IMAGES_SUBPATH,
                                          x_col=X_COL,
                                          y_col=Y_COL,
                                          target_size=TARGET_SIZE,
                                          batch_size=BATCH_SIZE,
                                          class_mode=CLASS_MODE,
                                          subset='validation',
                                          shuffle=True,
                                          seed=RANDOM_SEED)

    return valid_datagen_flow
    

def create_model(input_shape=(IMAGE_HEIGHT, IMAGE_WIDTH, 3),
                 weights='imagenet',
                 activation_functions='elu', learning_rate=0.00005,
                 loss='mse', metrics=['mae']):
    
    optimizer = Adam(learning_rate=learning_rate)
    
    backbone = ResNet152V2(include_top=False,
                           weights=weights,
                           input_shape=input_shape)
                                          
    model = Sequential()
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(units=1, activation=activation_functions))
    model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
    
    return model
    


def train_model(model, train_data, test_data, batch_size=None, epochs=15,
                steps_per_epoch=None, validation_steps=None):
    try:        
        model.fit(train_data,
                  validation_data=test_data,
                  batch_size=batch_size,
                  epochs=epochs,
                  steps_per_epoch=steps_per_epoch,
                  validation_steps=validation_steps,
                  verbose=2,
                  shuffle=True)
                                    
        model.save_weights('ResNet152V2_weights_real_age.h5')
                
        model = create_model(input_shape=(IMAGE_HEIGHT, IMAGE_WIDTH, 3),
                             weights=None,
                             activation_functions='relu', learning_rate=0.00001)
                                               
        model.load_weights('ResNet152V2_weights_real_age.h5')                         
                                         
        model.fit(train_data,
                  validation_data=test_data,
                  batch_size=batch_size,
                  epochs=15,
                  steps_per_epoch=steps_per_epoch,
                  validation_steps=validation_steps,
                  verbose=2,
                  shuffle=True)
                  
    except Exception as exception:
        print('Exception', exception)
        
    return model

```
