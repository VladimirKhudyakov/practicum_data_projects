**32. Обработаем пропуски в столбце `rating`.**

````python
# цикл for по строкам в датафрейме 'data', которые содержат пропуски в столбце 'rating'
for idx, row in data[data['rating'].isna()].iterrows():
    if not (data.loc[(data['name'] == row['name']) & ~(data['rating'].isna()), 'rating'].empty):
        data.loc[(data['name'] == row['name']) & (data['rating'].isna()), 'rating'] = \
        data.loc[(data['name'] == row['name']), 'rating'].dropna().min()
````

Создадим функцию отрисовки диаграмм рассеяния и вычисления корреляций между количеством продаж игр и отзывами, для всевозможных платформ, которые есть в датафрейме `data_filtered`.

````python
#
def dependency_check_on_other_platforms(platform_name):
    
    (data_filtered[data_filtered['platform'] == platform_name]
             .plot(x='user_score',
                   y='total_sales',
                   kind='scatter',
                   alpha=0.5,
                   title=f'Зависимость количества продаж игр \n от отзывов покупателей \n(для платформы {platform_name})',
                   xlabel='Отзывы покупателей',
                   ylabel='Количество продаж'))
    plt.show()
    
    correlation_coefficient = (data_filtered[data_filtered['platform'] == platform_name]['user_score']
    .corr(data_filtered[data_filtered['platform'] == platform_name]['total_sales']))
    
    print(f'Линейный коэффициент корреляции, между количеством продаж игр и отзывами покупателей,\nравен (платформа {platform_name}):', correlation_coefficient)

    (data_filtered[data_filtered['platform'] == platform_name]
             .plot(x='critic_score',
                   y='total_sales',
                   kind='scatter',
                   alpha=0.5,
                   title=f'Зависимость количества продаж игр \n от отзывов критиков \n(для платформы {platform_name})',
                   xlabel='Отзывы критиков',
                   ylabel='Количество продаж'))
    
    
    plt.show()
                   
    correlation_coefficient = (data_filtered[data_filtered['platform'] == platform_name]['critic_score']
    .corr(data_filtered[data_filtered['platform'] == platform_name]['total_sales']))
                   
    print(f'Линейный коэффициент корреляции, между количеством продаж игр и отзывами критиков,\nравен (платформа {platform_name}):', correlation_coefficient)
````

### Проверка гипотез

**48. Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**

```
H_0: Средние пользовательские рейтинги платформы Xbox One = cредним пользовательские рейтингам платформ PC
H_1: Средние пользовательские рейтинги платформы Xbox One ≠ cредним пользовательские рейтингам платформ PC
alpha = 0.05
```

````python
# вызываем метод для проверки гипотезы
results = st.ttest_ind(data_filtered[data_filtered['platform'] == 'XOne']['user_score'].dropna(),
                       data_filtered[data_filtered['platform'] == 'PC']['user_score'].dropna(),
                       equal_var= True)


# устанавливаем значение уровня статической значимости (1%)
alpha = 0.01 

# выводим значения p-value на экран 
print('p-value:',results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**49. Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные.**

```
H_0: Средние пользовательские рейтинги жанра Action = средним пользовательским рейтингам жанра Sports
H_1: Средние пользовательские рейтинги жанра Action ≠ средним пользовательским рейтингам жанра Sports
alpha = 0.05
```

````python
# вызываем метод для проверки гипотезы
results = st.ttest_ind(data_filtered[data_filtered['genre'] == 'Action']['user_score'].dropna(),
                       data_filtered[data_filtered['genre'] == 'Sports']['user_score'].dropna(),
                       equal_var= False)


# устанавливаем значение уровня статической значимости (1%)
alpha = 0.01 

# выводим значения p-value на экран 
print('p-value:',results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

**Вывод**

Наши гипотезы касаются средних значений генеральных совокупностей и звучат так: средние двух генеральных совокупностей равны между собой.

Сначала мы сформулировали нулевую гипотезу `H0`, согласно правилу формулирования гипотез, нулевая гипотеза формулируется так, чтобы использовать знак равенства `=` (среднее первой генеральной совокупности равно среднему второй генеральной совокупности). После, на основании гипотезы `H0`, мы сформулировали двухстороннюю гипотезу `H1`, со знаком `≠` (среднее первой генеральной совокупности не равно среднему второй генеральной совокупности).

Уровень статистической значимости мы выбрали `1%` (0.01) на основании того, что это самое минимальное значение, которое позволит более точно проверить нулевую гипотезу (повысить вероятность подтверждения гипотезы `H0`).


 - Первую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые - мы подтвердили. Вероятность получить различие средних рейтингов данных платформ не превысит 1%.
 - Вторую гипотезу: средние пользовательские рейтинги жанров Action и Sports разные - мы подтвердили. Вероятность получить равенство средних рейтингов данных жанров не превысит 1%.
